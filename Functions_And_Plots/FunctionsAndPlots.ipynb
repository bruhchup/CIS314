# Functions and Plots

```python
import matplotlib.pyplot as plt
import random
```

# Randomized scatter plot

```python
x = []
y = []

def plot_Random(x,y):
    '''draws a random scatter plot with randomized x values between 1-100. y = x^2'''
    for i in range(0,10):
        z = random.randint(1,100)
        x.append(z)

    for i in range(0,10):
        z = random.randint(1,100)
        y.append(x[i] ** 2)
        
    plt.scatter(x,y)
    plt.xlabel('x value')
    plt.ylabel('y value')
    plt.show()

plot_Random(x,y)
print(x, y)
```

# Example Functions

plot functions using variation in parameters

```python
x = [9.25,11,13.5,15,15.75]
y = [2.54,4.10,1.21,3.90,15.75]

def simplePlot_1(x,y):
    '''simple line graph'''
    plt.plot(x,y)
    plt.xlabel('x value')
    plt.ylabel('y value')
    plt.show()

def simplePlot_2(*,x,y):
    '''simple line graph. can only pass keyword arguments.'''
    plt.plot(x,y)
    plt.xlabel('x value')
    plt.ylabel('y value')
    plt.show()

def simplePlot_3(x,y,/):
    '''simple line graph. can only pass positional arguments.'''
    plt.plot(x,y)
    plt.xlabel('x value')
    plt.ylabel('y value')
    plt.show()
```

```python
#function without '*' or '/' parameters

simplePlot_1(x,y)
```

```python
#function with '*' parameter

simplePlot_2(x = x,y = y)
```

```python
#function with neither '/' or '*' parameters

simplePlot_3(x,y)
```

```python
#scatter plot

plt.scatter(x,y)
```
